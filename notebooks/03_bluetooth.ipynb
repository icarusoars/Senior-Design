```python
sensor_data = {
    'timestamp': [],
    'flex1'     : [],
    'flex2'     : [],
    'flex3'     : [],
    'flex4'     : [],
    'pres1'     : []
}
```

- found left index, found right index
    - process data
    - buffer = buffer[rb_idx+1:]
- did not find left index, did not find right index
    - do nothing
- found left index, did not find right index
    - do nothing
- did not find left index, found right index
    - throw away right index
    - buffer = buffer[rb_idx+1:]

```python
def process_buffer():

    global sensor_data
    global buffer
    
    lb_idx = buffer.find('{')
    rb_idx = buffer.find('}')
    
    while (lb_idx != -1 and rb_idx != -1):
        if lb_idx < rb_idx:
            data = json.loads(buffer[lb_idx:rb_idx+1])

            print("-----------------")
            # print(buffer)
            print(data)
            
            # record sensor data
            for k,v in data.items():
                sensor_data[k].append(v)
            
            buffer = buffer[rb_idx+1:]
        else:
            buffer = buffer[lb_idx:]
        
        lb_idx = buffer.find('{')
        rb_idx = buffer.find('}')
```

I ran below code for 9 seconds,
the timestamps went from 1548833 to 1557320 milliseconds, which is 8.5 seconds. Considering human error in stop-watch and other factors, the bluetooth is accurately receiving data at 100 millisecond intervals or at 10 Hz.

```python
bd_addr = "98:D3:11:FC:4F:48"
port = 1
sock = bluetooth.BluetoothSocket( bluetooth.RFCOMM )
sock.connect((bd_addr, port))
print('Connected\n')
sock.settimeout(0.5)


buffer = ""

count = 0;
while (count < 100000):
    try:
        data = sock.recv(4096)

        buffer += data.decode("utf-8")
        process_buffer()
    except Exception as e:
        print(e)
    
    time.sleep(2)
    count += 1


sock.close()
```
